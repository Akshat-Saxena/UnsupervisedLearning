```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("Country-data.csv")
```

```python
X = df.iloc[:,[1,5,9]].values  #independent 


Y = df.iloc[:,0].values  # dependent
```

```python
sc= StandardScaler()
X=sc.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python

df['ClusterID']= pd.Series(kmeans.labels_)
```

```python
df.head()
```

```python


GDPP=pd.DataFrame(df.groupby(["ClusterID"]).gdpp.mean())
child_mort=pd.DataFrame(df.groupby(["ClusterID"]).child_mort.mean())
income=pd.DataFrame(df.groupby(["ClusterID"]).income.mean())


print(GDPP)
print(child_mort)
print(income)
```

```python
comb_df = pd.concat([GDPP,child_mort,income], axis=1)
comb_df
```

### Countries with high gdpp and low child mortality are Developed countries
